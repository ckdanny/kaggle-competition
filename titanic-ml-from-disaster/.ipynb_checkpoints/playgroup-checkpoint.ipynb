````python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
print("You can find the data set in the below paths")
for dirname, _, filenames in os.walk(DATA_PATH):
    for filename in filenames:
        print("- ",os.path.join(dirname, filename))

````

````python
train_data=pd.read_csv("./data/titanic/train.csv")
# show first five rows of the table in train.csv
print("Below is the training data")
train_data.head()
````

````python
test_data=pd.read_csv("./data/titanic/test.csv")
# show first five rows of the table in train.csv
print("Below is the testing data")
test_data.head()
````

````python
def survive_percentage(sex:str)->float:
    assert sex in ['male','female'], 'invalid input parameter, male or female?'
    # train_data.loc -> select the row by the condition
    # train_data.loc[{condition}][{field}]
    filtered_sex = train_data.loc[train_data.Sex==sex]['Survived']
    # Survived is binary, i.e. 0 or 1
    rate = sum(filtered_sex)/len(filtered_sex)
    return round(rate, 3)*100

print(f'% of men survived: {survive_percentage(sex="male")}%')
print(f'% of women survived: {survive_percentage(sex="female")}%')
print("We can say that the gender is a strong indicator of survival")
````

````python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch"]

# `get_dummies is to convert the categorical data into binary data
# https://tinyurl.com/y2aks33e

# Convert process for both train data and test data
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])


# Build model
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)

predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
# output.to_csv('my_submission.csv', index=False)
# print("Your submission was successfully saved!")
output.head()
````

````python
feature_names = list(features)
# class_names = list(map(lambda i:str(i), model.classes_))
class_names = ["dead","live", "2"]
````

````python
estimator = model.estimators_[0]

from sklearn import tree
import matplotlib.pyplot as plt

fn= feature_names
cn = class_names
fig, axes = plt.subplots(nrows = 1,ncols = 5,figsize = (10,2), dpi=900)

# Only show the tree (estimator) from 0~1
for index in range(0, 5):
    tree.plot_tree(model.estimators_[index],
                   feature_names = fn, 
                   class_names=cn,
                   filled = True,
                   ax = axes[index]);
    axes[index].set_title('Estimator: ' + str(index), fontsize = 11)
````
